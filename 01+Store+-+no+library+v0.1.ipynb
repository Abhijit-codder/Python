#### Total sales for each category

```python
cat_sales = dict()
```

```python
for rec in raw_text:
    rec_split = rec.split(",")
    cat = rec_split[10]
    sales = float(rec_split[4])
    #if category entry exists in the lookup dictionary, append the sales value; create new entry otherwise
    if cat in cat_sales:
        cat_sales[cat] = cat_sales[cat] + sales
    else:
        cat_sales[cat] = sales
```
